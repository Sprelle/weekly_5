```python
import pandas as pd
import plotly as plt
import numpy as np
from pprint import pprint
```

```python
df = pd.read_excel("HSDPA_Traffic_ Dataset.xlsx")
```

```python
sites = np.array(df["Sites"])
times = np.array(df.columns[1:])
traffic = np.array(df)
traffic = np.delete(traffic, 0, 1)
traffic = traffic.astype(float)

traffic_max = dict()

for time_index, time in enumerate(times):
    week_number = time.date().isocalendar()[1]
    
    if week_number not in traffic_max:
        traffic_max[week_number] = dict()
    for site in sites:
        if site not in traffic_max[week_number]:
            traffic_max[week_number][site] = list()
            
for week in traffic_max.keys():
    for time_nr, time in enumerate(times):
        #print(time)
        week_number = time.date().isocalendar()[1]
        if week_number != week:
            continue
        for site_nr, site in enumerate(sites):
            traffic_max[week_number][site].append(traffic[site_nr][time_nr])

for week in traffic_max:
    for site in traffic_max[week]:
        traffic_max[week][site] = np.max(traffic_max[week][site])


#print(sites)
#print(times)
print(traffic_max[53])

plottable = np.zeros((len(sites), len(traffic_max.keys())))
print(plottable.shape)

for week_index, week in enumerate(traffic_max.keys()):
    #print(week_index, week)
    
    for site_index, site in enumerate(traffic_max[week].keys()):
        #print(site_index, site)
        
        plottable[site_index][week_index] = traffic_max[week][site]
        
print(plottable)
```

```python
import plotly.graph_objects as go
print(list(traffic_max.keys()))

fig = go.Figure(data=go.Heatmap(
                   z=plottable,
                   x=["Week 53","Week 1","Week 2","Week 3","Week 4",],
                   y=sites,
            colorscale='thermal',
            
                    
                                ))
fig.update_layout(
    title= "Max traffic recorded (MB) for sites"
                    )
fig.show()
```
